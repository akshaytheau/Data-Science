Clustering is a technique used in machine learning and data mining to group similar objects or data points together based on their characteristics or patterns. It is an unsupervised learning method, meaning it does not require labeled data or predefined categories.

The goal of clustering is to identify inherent structures or relationships within a dataset and group similar items together while keeping dissimilar items separate. Clustering algorithms accomplish this by quantifying the similarity or distance between data points and creating clusters based on those similarities.

There are various clustering algorithms available, each with its own approach to grouping data points. Some popular clustering algorithms include k-means, hierarchical clustering, DBSCAN (Density-Based Spatial Clustering of Applications with Noise), and Gaussian mixture models.

Clustering finds applications in various fields such as data analysis, image segmentation, customer segmentation, document clustering, anomaly detection, and many more. It is often used as a preliminary step in data exploration and as a tool to discover hidden patterns or groupings within datasets.





```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data_X,data_y=make_blobs(n_samples=500,centers=4,n_features=2,random_state=14)
```

```python
plt.scatter(data_X[:,0],data_X[:,1])
```

```python
from sklearn.cluster import KMeans
wcss=[]
for k in range(1,11):
  kmeans=KMeans(n_clusters=k,init='k-means++')
  kmeans.fit(data_X)
  wcss.append(kmeans.inertia_)
```

'random' : centriods will be selected randomly.
'k-means++' : centriods will be selected based on empirical probability distribution of the points contribution to the overall inertia

```python
wcss
```

```python
## plot elbow curve
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,init="k-means++")
```

```python
y_labels=kmeans.fit_predict(data_X)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_coefficients=[]
for k in range(2,11):
  kmeans=KMeans(n_clusters=k,init="k-means++")
  kmeans.fit(data_X)
  score=silhouette_score(data_X,kmeans.labels_)
  silhouette_coefficients.append(score)
```

```python
silhouette_coefficients
```

```python
plt.plot(range(2,11),silhouette_coefficients)
plt.xticks(range(2,11))
plt.xlabel("Number of Cluters")
plt.ylabel("Silhoutte Coeffecient")
plt.show()
```


The silhouette score is a metric used to evaluate the quality of clustering results. It measures how well each data point fits into its assigned cluster and how distinct the clusters are from each other.

The silhouette score is calculated for each data point and ranges from -1 to 1. A higher score indicates better clustering results:

A score close to +1 suggests that the data point is well-matched to its assigned cluster and is far away from the neighboring clusters.
A score close to 0 suggests that the data point is near the decision boundary between two clusters.
A score close to -1 indicates that the data point may have been assigned to the wrong cluster.
To calculate the silhouette score, the following steps are generally followed:

For each data point, compute two measures:

"a": The average distance between the data point and other points within the same cluster (intra-cluster distance).
"b": The average distance between the data point and points in the nearest neighboring cluster (inter-cluster distance).
Compute the silhouette coefficient for each data point using the formula:
silhouette_coefficient = (b - a) / max(a, b)

Calculate the average silhouette coefficient over all data points to obtain the silhouette score for the clustering result.

The silhouette score provides an indication of the compactness and separation of clusters. Higher scores imply better-defined clusters with less overlap.

It's important to note that the silhouette score is just one of several metrics available for evaluating clustering results. Depending on the specific context and data characteristics, other metrics such as the Calinski-Harabasz index or Davies-Bouldin index may also be used.
