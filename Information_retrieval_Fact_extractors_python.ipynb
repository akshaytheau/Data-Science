```python
import spacy
import wikipedia

nlp = spacy.load('en_core_web_sm')

subject = 'Anand Mahindra'

wikiResults = wikipedia.search(subject)
wikiPage = wikipedia.page(wikiResults[0]).content

doc = nlp(wikiPage)

for entity in doc.ents:
    print(f"{entity.text} ({entity.label_})")
```

```python
import wikipedia
import spacy
import textacy
import en_core_web_sm

subject = 'Anand Mahindra'

wikiResults = wikipedia.search(subject)
print(len(wikiResults))
wikiPage = wikipedia.page(wikiResults[0]).content
print("\n\nwikiPage:", wikiPage, "'\n")
nlp = en_core_web_sm.load()
document = nlp(wikiPage)
uniqueStatements = list()
for word in ["Mahindra","Anand"]:
    for cue in ["be", "have", "write", "talk", "talk about"]:
        statments = textacy.extract.semistructured_statements(document,entity =  word, cue = cue )
        for statement in statments:
            uniqueStatements.append(statement)

print("found", len(uniqueStatements), "statements.")
for statement in uniqueStatements:
    entity, cue, fact = statement
    fact = map(str,fact)
    print("* entity:",entity, ", cue:", cue, ", fact:", ' '.join(fact))
```
