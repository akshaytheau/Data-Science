```python
def sigmoid(x):
  z = np.exp(-x)
  return 1/(1+z)
```

```python
print(sigmoid(0))
print(sigmoid(2))
print(sigmoid(4))
print(sigmoid(10))
print(sigmoid(100))
print(sigmoid(1000))
```

```python
print(sigmoid(0))
print(sigmoid(-2))
print(sigmoid(-4))
print(sigmoid(-10))
print(sigmoid(-100))
print(sigmoid(-1000))
```

```python
def sigmoid_positive(x):
  z = np.exp(-x)
  return 1/(1+z)

def sigmoid_negative(x):
  z = np.exp(x)
  return z/(1+z)
```

```python
print(sigmoid_positive(0))
print(sigmoid_positive(10))
print(sigmoid_positive(100))
print(sigmoid_positive(1000))
print(sigmoid_positive(1000000000))
```

```python
print(sigmoid_negative(0))
print(sigmoid_negative(-10))
print(sigmoid_negative(-100))
print(sigmoid_negative(-1000))
print(sigmoid_negative(-1000000000))
```
